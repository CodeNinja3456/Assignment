```python
import pandas
```

```python
medical=pandas.read_csv('Medical_Reviews.csv')
medical.head(5)
```

```python
medical.isnull().sum()
```

```python
medical['Diagnosis Category'].fillna('Unknown',inplace=True)
medical['Diagnosis Sub Category'].fillna('Unknown',inplace=True)
medical['Treatment Category'].fillna('Unknown',inplace=True)
medical['Treatment Sub Category'].fillna('Unknown',inplace=True)
medical['Patient Gender'].fillna('Unknown',inplace=True)
medical['Age Range'].fillna('Unknown',inplace=True)
medical['Findings'].fillna('Unknown',inplace=True)
```

```python
medical.drop_duplicates().head(5)
```

```python
medical.groupby(['Diagnosis Category','Diagnosis Sub Category','Patient Gender']).count().head(5)
```

```python
import pandas
```

```python
sample=pandas.read_csv('sample_texts.csv')
sample.head(5)
```

```python
sample.isnull().sum()
```

```python
X=sample['description'].values
```

```python
from sentence_transformers import SentenceTransformer
model =SentenceTransformer('distilbert-base-nli-stsb-mean-tokens')

embeddings=model.encode(X,show_progress_bar=True)
embeddings
```

```python
new_data=pandas.DataFrame(embeddings)
new_data
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_data=cosine_similarity(new_data)

cosine_data_table=pandas.DataFrame(cosine_data)
cosine_data_table
```

```python
def recommenders(id):
    index=cosine_data_table.loc[id].sort_values(ascending=False).index.tolist()[1:10]
    print ('Similar Text Articles are:',index)
    news_titles=sample['description'].loc[index].values
    print('Similar Text Article are:',news_titles)

recommenders(5)
```

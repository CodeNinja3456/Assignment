```python
import nltk
```

```python
text =  '''best, good, better, nice, connectivity, connection, connecting I just wanted to find some really 
cool new places such as Seattle in November. I’ve never visited before but no luck here. Some of these 
suggestions are just terrible... I had to laugh! Most suggestions were just your typical big cities, 
restaurants and bars. Nothing off the beaten path here. I don’t want to go these places 
for fun. Totally not worth getting this, This was such a beautiful book. I wasn’t even planning any travel 
when I came across this and just started flipping through the pages. I really like the cover and all the large 
glossy photographs in this book. John Smith did a wonderful job with the photography. I’ve found a 
perfect home for this on my coffee table. I’m planning a trip to Paris and Barcelona soon and I know this 
will come in handy. In the meantime, it’s perfect for assisting this armchair traveler, As a traveler, I really 
appreciated reading about these great places to visit. The author takes you all over the world. Even with 
all the free information online these days, I find I’m taking this book with me wherever I go and using it 
to discover hidden gem, this product was not good, i love this product, this product was not as expected, i 
like this product was nice to use'''    
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
token_words=word_tokenize(text)
token_words
```

```python
from nltk.corpus import stopwords
stopwords=set(stopwords.words('english'))
stopwords
```

```python
filtered_words=[]
for word in token_words:
     if word.lower() not in stopwords:
        filtered_words.append(word)
print("The new words without the common words is:",filtered_words) 
```

```python
new_words=[word for word in filtered_words if word.isalpha()]    
new_words
```

```python
from nltk.probability import FreqDist
frequency = FreqDist(new_words)
frequency.most_common(30)
```

```python

from wordcloud import WordCloud, STOPWORDS  
from PIL import Image 



def create_wordcloud(text):
        stopwords = set(STOPWORDS)
        wc = WordCloud(background_color="white",
        stopwords=stopwords,repeat=True) 
        wc.generate(str(text))
        wc.to_file("wc.png") 
        print("Word Cloud Saved Successfully")
        path="wc.png" 
        display(Image.open(path))
```
